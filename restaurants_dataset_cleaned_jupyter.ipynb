# 🍽️ Restaurants Dataset — Cleaning & Feature Engineering
---
**Dataset:** `restaurants.csv`  
**Goal:** Validate restaurant data, check quality, and create performance-based features.  
**Final output:** `restaurants_cleaned.csv`

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

## 2. Load Dataset

```python
# Load the raw restaurants CSV file
df = pd.read_csv("restaurants.csv")

# Quick shape check
print("Shape:", df.shape)
df.head()
```

## 3. Explore the Data

```python
df.info()
```

```python
df.describe()
```

## 4. Check Duplicates & Missing Values

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

## 5. Data Validation

```python
invalid_rating = df[(df['rating'] < 1) | (df['rating'] > 5)]
print(f"Invalid ratings (out of 1-5): {len(invalid_rating)}")
```

```python
invalid_prep = df[df['avg_prep_time'] <= 0]
print(f"Invalid prep time rows: {len(invalid_prep)}")
```

```python
invalid_cost = df[df['avg_cost_for_two'] <= 0]
print(f"Invalid cost rows: {len(invalid_cost)}")
```

```python
print("\nrestaurant_type unique:", df['restaurant_type'].unique())
print("active_status unique   :", df['active_status'].unique())
```

## 6. Feature Engineering

```python
# Restaurants with rating >= 4.5 are considered high rated
df['high_rated'] = np.where(df['rating'] >= 4.5,'Yes', 'No')
```

```python
# avg_cost_for_two > 1500 is considered an expensive/premium restaurant
df['expensive_restaurant'] = np.where(df['avg_cost_for_two'] > 1500,'Yes','No')
```

```python
# Restaurants that prepare orders within 20 minutes
df['fast_prep'] = np.where(df['avg_prep_time'] <= 20,'Yes','No')
```

## 7. Final Check & Export

```python
# Final shape and null check before saving
print("Final Shape:", df.shape)
print("\nNull values after all cleaning steps:")
print(df.isnull().sum())
df.head()
```

```python
# Save cleaned dataset to CSV
df.to_csv("restaurants_cleaned.csv", index=False)
print("✅ restaurants_cleaned.csv saved successfully!")
```
