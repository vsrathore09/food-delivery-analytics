```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
orders = pd.read_csv("orders_cleaned.csv")

customers = pd.read_csv(
    "customers_cleaned.csv"
)

restaurants = pd.read_csv(
    "restaurants_cleaned.csv"
)

delivery = pd.read_csv(
    "delivery_cleaned.csv"
)
```

```python
engine = create_engine(
    "mysql+pymysql://root:Password@localhost/food_delivery_analysis"
)
```

```python
orders.to_sql(
    name='orders',
    con=engine,
    if_exists='replace',
    index=False
)
```

```python
customers.to_sql(
    name='customers',
    con=engine,
    if_exists='replace',
    index=False
)
```

```python
restaurants.to_sql(
    name='restaurants',
    con=engine,
    if_exists='replace',
    index=False
)
```

```python
delivery.to_sql(
    name='delivery',
    con=engine,
    if_exists='replace',
    index=False
)
```
